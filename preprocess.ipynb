## Explanation: Why 4 Tables for Data Cleaning vs 7 Tables in ER Diagram

### Understanding the Data Architecture

**ER Diagram Structure (7 Tables):**
According to the project documentation, the ER diagram specifies 7 entities:
1. **Users** - Authentication and user management
2. **Online Shops** - Vendor/business information
3. **Campaigns** - Marketing campaign data
4. **Reports** - Generated analytics reports
5. **PredictionModels** - ML model metadata
6. **Dashboards** - User dashboard configurations
7. **Sales/Transaction Data** - The actual sales transactions (implied)

**Why We Used 4 Logical Tables for Data Cleaning:**

In our data cleaning process, we organized the data into 4 logical tables/conceptual groups:

1. **Sales Transactions Table** (Primary Data Source)
   - Contains: Invoice Date, Invoice ID, Customer Name, Total Sales, Product Name, Product Category
   - Purpose: Core transactional data for analysis
   - Source: `sales_in_saudi_arabia.xlsx`

2. **Customer Attributes Table** (Derived from Sales Data)
   - Contains: Customer Name, Customer Type, Customer Gender, City, Customer Satisfaction
   - Purpose: Customer demographic and behavioral attributes
   - Source: Extracted from sales transactions

3. **Channel & Employee Table** (Operational Data)
   - Contains: Channel (Store/Online), Employee Name, Manager Name
   - Purpose: Operational and channel performance analysis
   - Source: Extracted from sales transactions

4. **Product & Category Table** (Product Data)
   - Contains: Product Name, Product Category
   - Purpose: Product performance and category analysis
   - Source: Extracted from sales transactions

**Key Difference:**
- The ER diagram represents the **full system architecture** including user management, campaigns, reports, and dashboards
- Our data cleaning focused on the **transactional sales data** which is the foundation for all marketing intelligence
- The other 3 tables (Users, Campaigns, Reports) are system-level tables that don't require the same data cleaning process
- We extracted 4 logical data groups from the single sales file for focused cleaning and analysis

**Data Flow:**
```
Raw Sales Data (1 file) 
    ↓
Data Cleaning (4 logical tables)
    ↓
Feature Engineering (32 features)
    ↓
ML-Ready Dataset (25 features)
    ↓
Integration with System Tables (Users, Campaigns, Reports)
```

This approach ensures we have clean, structured data ready for marketing intelligence while maintaining alignment with the overall system architecture.

## Step 4: Feature Engineering for RevXcel KPIs
### Calculate Customer Lifetime Value (CLV), RFM metrics, and other marketing KPIs


````python
# Feature Engineering for Marketing Intelligence
print("=" * 70)
print("FEATURE ENGINEERING FOR REVXCEL KPIs")
print("=" * 70)

df_features = df_clean.copy()

# 1. Extract temporal features from Invoice Date
if 'Invoice Date' in df_features.columns:
    df_features['Year'] = df_features['Invoice Date'].dt.year
    df_features['Month'] = df_features['Invoice Date'].dt.month
    df_features['Day'] = df_features['Invoice Date'].dt.day
    df_features['DayOfWeek'] = df_features['Invoice Date'].dt.dayofweek
    df_features['DayName'] = df_features['Invoice Date'].dt.day_name()
    df_features['MonthName'] = df_features['Invoice Date'].dt.month_name()
    df_features['Quarter'] = df_features['Invoice Date'].dt.quarter
    print("✓ Extracted temporal features from Invoice Date")

# 2. Calculate Customer-Level Metrics (for CLV calculation)
if 'Customer Name' in df_features.columns and 'Total Sales' in df_features.columns:
    customer_metrics = df_features.groupby('Customer Name').agg({
        'Total Sales': ['sum', 'mean', 'count'],
        'Invoice Date': ['min', 'max']
    }).reset_index()
    
    customer_metrics.columns = ['Customer Name', 'Total_Revenue', 'Avg_Transaction_Value', 
                                'Transaction_Count', 'First_Purchase_Date', 'Last_Purchase_Date']
    
    # Calculate customer lifetime (days since first purchase)
    customer_metrics['Customer_Lifetime_Days'] = (
        customer_metrics['Last_Purchase_Date'] - customer_metrics['First_Purchase_Date']
    ).dt.days
    customer_metrics['Customer_Lifetime_Days'] = customer_metrics['Customer_Lifetime_Days'].fillna(0)
    
    # Calculate Customer Lifetime Value (CLV)
    # CLV = Total Revenue / (Customer Lifetime in years + 1) to avoid division by zero
    customer_metrics['CLV'] = customer_metrics['Total_Revenue'] / (
        (customer_metrics['Customer_Lifetime_Days'] / 365.25) + 1
    )
    
    # Calculate Purchase Frequency (transactions per year)
    customer_metrics['Purchase_Frequency'] = customer_metrics['Transaction_Count'] / (
        (customer_metrics['Customer_Lifetime_Days'] / 365.25) + 1
    )
    
    # Merge back to main dataframe
    df_features = df_features.merge(
        customer_metrics[['Customer Name', 'Total_Revenue', 'Avg_Transaction_Value', 
                          'Transaction_Count', 'CLV', 'Purchase_Frequency']],
        on='Customer Name',
        how='left'
    )
    print("✓ Calculated Customer Lifetime Value (CLV) and related metrics")

# 3. RFM Analysis (Recency, Frequency, Monetary)
if 'Customer Name' in df_features.columns and 'Invoice Date' in df_features.columns:
    # Get the most recent date in dataset
    max_date = df_features['Invoice Date'].max()
    
    rfm = df_features.groupby('Customer Name').agg({
        'Invoice Date': lambda x: (max_date - x.max()).days,  # Recency
        'Invoice Date': 'count',  # Frequency
        'Total Sales': 'sum'      # Monetary
    }).reset_index()
    
    # Fix the aggregation (pandas issue with same column twice)
    rfm_calc = df_features.groupby('Customer Name').agg({
        'Invoice Date': ['max', 'count'],
        'Total Sales': 'sum'
    }).reset_index()
    rfm_calc.columns = ['Customer Name', 'Last_Purchase', 'Frequency', 'Monetary']
    rfm_calc['Recency'] = (max_date - rfm_calc['Last_Purchase']).dt.days
    
    # Create RFM scores (1-5 scale) with robust handling of edge cases
    def create_rfm_score(series, reverse=False):
        """
        Create RFM score (1-5) using percentile-based approach
        reverse=True for Recency (higher recency = lower score)
        """
        # Remove NaN values for calculation
        series_clean = series.dropna()
        if len(series_clean) == 0:
            return pd.Series([3] * len(series), index=series.index, dtype=int)  # Default to middle score
        
        # Use percentile-based approach with pd.cut for more control
        try:
            # Calculate percentiles
            percentiles = [0, 20, 40, 60, 80, 100]
            bins = [series_clean.quantile(p/100) for p in percentiles]
            
            # Remove duplicate bin edges and ensure we have at least 2 bins
            bins = sorted(list(set(bins)))
            if len(bins) < 2:
                bins = [series_clean.min() - 0.01, series_clean.max() + 0.01]
            
            # Determine number of actual bins
            n_bins = len(bins) - 1
            n_bins = min(5, n_bins)  # Cap at 5
            
            # Create labels based on number of bins and direction
            if reverse:
                # For Recency: higher recency (more days) = lower score
                labels = list(range(n_bins, 0, -1))  # [5,4,3,2,1] or fewer
            else:
                # For Frequency/Monetary: higher value = higher score
                labels = list(range(1, n_bins + 1))  # [1,2,3,4,5] or fewer
            
            # Use pd.cut with custom bins
            scores = pd.cut(series, bins=bins, labels=labels[:n_bins], 
                           include_lowest=True, duplicates='drop')
            
            # Convert to numeric and handle any NaN
            scores = pd.to_numeric(scores, errors='coerce')
            scores = scores.fillna(3).astype(int)  # Fill NaN with middle score
            
            # Ensure scores are in valid range (1-5)
            scores = scores.clip(1, 5)
            return scores
            
        except Exception as e:
            # Fallback: use simple ranking approach
            try:
                # Use rank to create unique values, then assign to 5 groups
                if reverse:
                    ranked = series.rank(method='first', ascending=False, na_option='keep')
                else:
                    ranked = series.rank(method='first', ascending=True, na_option='keep')
                
                # Calculate quantiles on ranked values
                max_rank = ranked.max()
                if pd.isna(max_rank) or max_rank == 0:
                    return pd.Series([3] * len(series), index=series.index, dtype=int)
                
                # Create 5 groups based on rank percentiles
                if reverse:
                    scores = pd.cut(ranked, bins=[0, max_rank*0.2, max_rank*0.4, max_rank*0.6, max_rank*0.8, max_rank+1],
                                   labels=[5,4,3,2,1], include_lowest=True)
                else:
                    scores = pd.cut(ranked, bins=[0, max_rank*0.2, max_rank*0.4, max_rank*0.6, max_rank*0.8, max_rank+1],
                                   labels=[1,2,3,4,5], include_lowest=True)
                
                scores = pd.to_numeric(scores, errors='coerce').fillna(3).astype(int).clip(1, 5)
                return scores
            except:
                # Ultimate fallback: simple division into 5 groups
                if reverse:
                    # Higher recency = lower score
                    sorted_idx = series.sort_values(ascending=False).index
                else:
                    # Higher value = higher score
                    sorted_idx = series.sort_values(ascending=True).index
                
                n = len(sorted_idx)
                scores = pd.Series([3] * len(series), index=series.index, dtype=int)
                
                if n > 0:
                    chunk_size = max(1, n // 5)
                    for i in range(5):
                        start_idx = i * chunk_size
                        end_idx = (i + 1) * chunk_size if i < 4 else n
                        if reverse:
                            score_value = 5 - i
                        else:
                            score_value = i + 1
                        scores.loc[sorted_idx[start_idx:end_idx]] = score_value
                
                return scores
    
    # Apply scoring with error handling
    rfm_calc['R_Score'] = create_rfm_score(rfm_calc['Recency'], reverse=True)
    rfm_calc['F_Score'] = create_rfm_score(rfm_calc['Frequency'], reverse=False)
    rfm_calc['M_Score'] = create_rfm_score(rfm_calc['Monetary'], reverse=False)
    
    rfm_calc['RFM_Score'] = (
        rfm_calc['R_Score'].astype(int) + 
        rfm_calc['F_Score'].astype(int) + 
        rfm_calc['M_Score'].astype(int)
    )
    
    # Customer Segmentation based on RFM
    def segment_customer(rfm_score):
        if rfm_score >= 12:
            return 'Champions'
        elif rfm_score >= 9:
            return 'Loyal Customers'
        elif rfm_score >= 6:
            return 'Potential Loyalists'
        elif rfm_score >= 4:
            return 'At Risk'
        else:
            return 'Lost'
    
    rfm_calc['RFM_Segment'] = rfm_calc['RFM_Score'].apply(segment_customer)
    
    # Merge RFM data
    df_features = df_features.merge(
        rfm_calc[['Customer Name', 'Recency', 'Frequency', 'Monetary', 
                 'R_Score', 'F_Score', 'M_Score', 'RFM_Score', 'RFM_Segment']],
        on='Customer Name',
        how='left'
    )
    print("✓ Calculated RFM (Recency, Frequency, Monetary) analysis and customer segmentation")

# 4. Channel Performance Metrics
if 'Channel' in df_features.columns:
    channel_metrics = df_features.groupby('Channel').agg({
        'Total Sales': ['sum', 'mean', 'count'],
        'Invoice Date': 'nunique'
    }).reset_index()
    channel_metrics.columns = ['Channel', 'Channel_Total_Sales', 'Channel_Avg_Sales', 
                               'Channel_Transaction_Count', 'Channel_Unique_Dates']
    print("✓ Calculated channel performance metrics")

# 5. Product Category Performance
if 'Product Category' in df_features.columns:
    category_metrics = df_features.groupby('Product Category').agg({
        'Total Sales': ['sum', 'mean', 'count']
    }).reset_index()
    category_metrics.columns = ['Product Category', 'Category_Total_Sales', 
                               'Category_Avg_Sales', 'Category_Transaction_Count']
    print("✓ Calculated product category performance metrics")

print(f"\n✓ Feature engineering complete. New shape: {df_features.shape}")
print("=" * 70)
````

````python
# Personal Customer Behavior Extraction
print("=" * 70)
print("PERSONAL CUSTOMER BEHAVIOR EXTRACTION")
print("=" * 70)

print("\n" + "=" * 70)
print("HOW WE EXTRACT PERSONAL CUSTOMER BEHAVIOR")
print("=" * 70)
print("""
Our data analysis enables extraction of individual customer behavior patterns through:

1. TEMPORAL BEHAVIOR ANALYSIS
   - Purchase frequency patterns (daily, weekly, monthly, yearly)
   - Time between purchases (e.g., customer buys every 30 days)
   - Seasonal purchase patterns (e.g., customer buys more in winter)
   - Day of week preferences (e.g., customer shops on weekends)

2. SPENDING BEHAVIOR ANALYSIS
   - Price range preferences (e.g., customer typically spends $500-$1000)
   - Spending consistency (consistent vs variable spender)
   - Price sensitivity (narrow vs wide price range acceptance)
   - Transaction value trends over time

3. PRODUCT PREFERENCE ANALYSIS
   - Preferred product categories
   - Brand loyalty patterns
   - Product switching behavior
   - Cross-category purchase patterns

4. CHANNEL BEHAVIOR ANALYSIS
   - Preferred shopping channel (Online vs Store)
   - Multi-channel usage patterns
   - Channel switching behavior
   - Channel-specific spending patterns

5. LIFECYCLE BEHAVIOR ANALYSIS
   - Customer journey stage (New → Loyal → At Risk → Lost)
   - Behavior changes over customer lifetime
   - Engagement level trends
   - Retention probability
""")

# Extract detailed personal behavior for sample customers
print("\n" + "=" * 70)
print("DETAILED PERSONAL BEHAVIOR EXAMPLES")
print("=" * 70)

if 'Customer Name' in df_features.columns:
    # Get top 5 customers for detailed analysis
    top_customers = df_features.groupby('Customer Name')['Total Sales'].sum().nlargest(5).index
    
    for customer in top_customers[:3]:  # Show first 3 as examples
        customer_data = df_features[df_features['Customer Name'] == customer].copy()
        customer_data = customer_data.sort_values('Invoice Date')
        
        print(f"\n{'='*70}")
        print(f"CUSTOMER: {customer}")
        print(f"{'='*70}")
        
        # Basic Stats
        total_spent = customer_data['Total Sales'].sum()
        transaction_count = len(customer_data)
        avg_transaction = customer_data['Total Sales'].mean()
        first_purchase = customer_data['Invoice Date'].min()
        last_purchase = customer_data['Invoice Date'].max()
        customer_lifetime = (last_purchase - first_purchase).days
        
        print(f"\n1. BASIC STATISTICS:")
        print(f"   Total Spent: ${total_spent:,.2f}")
        print(f"   Number of Transactions: {transaction_count}")
        print(f"   Average Transaction Value: ${avg_transaction:,.2f}")
        print(f"   First Purchase: {first_purchase.strftime('%Y-%m-%d')}")
        print(f"   Last Purchase: {last_purchase.strftime('%Y-%m-%d')}")
        print(f"   Customer Lifetime: {customer_lifetime} days")
        
        # Purchase Frequency Pattern
        if customer_lifetime > 0:
            avg_days_between = customer_lifetime / transaction_count if transaction_count > 1 else customer_lifetime
            print(f"\n2. PURCHASE FREQUENCY PATTERN:")
            print(f"   Average days between purchases: {avg_days_between:.1f} days")
            if avg_days_between < 30:
                print(f"   → Behavior: Frequent buyer (purchases every {avg_days_between:.0f} days)")
            elif avg_days_between < 90:
                print(f"   → Behavior: Regular buyer (purchases every {avg_days_between:.0f} days)")
            elif avg_days_between < 180:
                print(f"   → Behavior: Occasional buyer (purchases every {avg_days_between:.0f} days)")
            else:
                print(f"   → Behavior: Infrequent buyer (purchases every {avg_days_between:.0f} days)")
        
        # Spending Pattern
        min_spend = customer_data['Total Sales'].min()
        max_spend = customer_data['Total Sales'].max()
        std_spend = customer_data['Total Sales'].std()
        cv = (std_spend / avg_transaction) * 100 if avg_transaction > 0 else 0
        
        print(f"\n3. SPENDING PATTERN:")
        print(f"   Spending Range: ${min_spend:,.2f} - ${max_spend:,.2f}")
        print(f"   Spending Variability (CV): {cv:.1f}%")
        if cv < 30:
            print(f"   → Behavior: Consistent spender (predictable spending)")
        elif cv < 60:
            print(f"   → Behavior: Moderate spender (some variability)")
        else:
            print(f"   → Behavior: Variable spender (high variability)")
        
        # Price Sensitivity
        price_range = max_spend - min_spend
        if price_range < 500:
            print(f"   → Price Sensitivity: Price-sensitive (narrow range: ${price_range:,.2f})")
        elif price_range > 3000:
            print(f"   → Price Sensitivity: Price-flexible (wide range: ${price_range:,.2f})")
        else:
            print(f"   → Price Sensitivity: Moderate (range: ${price_range:,.2f})")
        
        # Channel Preference
        if 'Channel' in customer_data.columns:
            channel_dist = customer_data['Channel'].value_counts()
            preferred_channel = channel_dist.index[0]
            channel_pct = (channel_dist.iloc[0] / len(customer_data)) * 100
            
            print(f"\n4. CHANNEL PREFERENCE:")
            print(f"   Preferred Channel: {preferred_channel} ({channel_pct:.1f}% of transactions)")
            if len(channel_dist) > 1:
                print(f"   → Behavior: Multi-channel customer")
            else:
                print(f"   → Behavior: Single-channel loyalist")
        
        # Product Category Preference
        if 'Product Category' in customer_data.columns:
            category_dist = customer_data['Product Category'].value_counts()
            preferred_category = category_dist.index[0]
            category_pct = (category_dist.iloc[0] / len(customer_data)) * 100
            
            print(f"\n5. PRODUCT CATEGORY PREFERENCE:")
            print(f"   Preferred Category: {preferred_category} ({category_pct:.1f}% of purchases)")
            print(f"   Top 3 Categories:")
            for idx, (cat, count) in enumerate(category_dist.head(3).items(), 1):
                print(f"     {idx}. {cat}: {count} purchases")
        
        # Day of Week Preference
        if 'DayName' in customer_data.columns:
            day_dist = customer_data['DayName'].value_counts()
            preferred_day = day_dist.index[0]
            
            print(f"\n6. SHOPPING TIME PREFERENCE:")
            print(f"   Preferred Shopping Day: {preferred_day}")
            print(f"   → Behavior: Customer prefers shopping on {preferred_day}s")
        
        # Customer Segment
        if 'RFM_Segment' in customer_data.columns:
            segment = customer_data['RFM_Segment'].iloc[0]
            print(f"\n7. CUSTOMER SEGMENT:")
            print(f"   Current Segment: {segment}")
            if segment == 'Champions':
                print(f"   → Behavior: High-value, frequent, recent buyer - VIP customer")
            elif segment == 'Loyal Customers':
                print(f"   → Behavior: Regular, loyal customer - maintain relationship")
            elif segment == 'Potential Loyalists':
                print(f"   → Behavior: Growing customer - nurture for loyalty")
            elif segment == 'At Risk':
                print(f"   → Behavior: Declining engagement - win-back needed")
            else:
                print(f"   → Behavior: Inactive customer - re-engagement required")
        
        # CLV
        if 'CLV' in customer_data.columns:
            clv = customer_data['CLV'].iloc[0]
            print(f"\n8. CUSTOMER LIFETIME VALUE:")
            print(f"   CLV: ${clv:,.2f}")
            if clv > 1000000:
                print(f"   → Value: Premium customer - high investment priority")
            elif clv > 500000:
                print(f"   → Value: High-value customer - significant investment")
            else:
                print(f"   → Value: Standard customer - standard marketing")

print("\n" + "=" * 70)
print("BEHAVIOR EXTRACTION METHODOLOGY")
print("=" * 70)
print("""
How We Extract Personal Behavior:

1. INDIVIDUAL TRANSACTION ANALYSIS
   - Analyze each customer's complete transaction history
   - Identify patterns in timing, amount, products, channels
   - Calculate statistical measures (mean, std, min, max)

2. TEMPORAL PATTERN RECOGNITION
   - Calculate time between purchases
   - Identify seasonal patterns
   - Detect day-of-week preferences
   - Recognize purchase frequency trends

3. BEHAVIORAL SEGMENTATION
   - Classify spending consistency (consistent vs variable)
   - Identify price sensitivity levels
   - Determine channel loyalty
   - Assess product category preferences

4. PREDICTIVE BEHAVIOR MODELING
   - Use RFM analysis for lifecycle stage
   - Calculate CLV for value assessment
   - Predict future purchase likelihood
   - Estimate churn risk

5. PERSONALIZED INSIGHTS GENERATION
   - Combine all patterns into customer profile
   - Generate actionable marketing recommendations
   - Create personalized campaign suggestions
   - Develop retention strategies

Example: "Customer X buys after a year at a specific price"
- We identify customers with ~365 day purchase intervals
- We note their typical price range
- We predict next purchase timing
- We recommend targeted campaigns at predicted time
- We suggest products in their price range
""")

print("\n" + "=" * 70)
print("✓ Personal customer behavior extraction complete")
print("=" * 70)
````

## Step 6: Data Preprocessing for Machine Learning
### Prepare data for ML models (encoding, scaling, feature selection)


````python
# Final preprocessing for ML models
print("=" * 70)
print("DATA PREPROCESSING FOR MACHINE LEARNING")
print("=" * 70)

df_ml = df_features.copy()

# 1. Encode categorical variables
print("\n1. Categorical Encoding:")
label_encoders = {}

categorical_cols = ['Channel', 'Customer Type', 'Product Category', 'City', 
                   'Customer Gender', 'Customer Satisfaction', 'RFM_Segment']

for col in categorical_cols:
    if col in df_ml.columns:
        le = LabelEncoder()
        df_ml[f'{col}_encoded'] = le.fit_transform(df_ml[col].astype(str))
        label_encoders[col] = le
        print(f"  ✓ Encoded {col} ({df_ml[col].nunique()} unique values)")

# 2. Select features for ML models
feature_columns = []

# Numerical features
numerical_features = ['Total Sales', 'Year', 'Month', 'Day', 'DayOfWeek', 'Quarter',
                     'Total_Revenue', 'Avg_Transaction_Value', 'Transaction_Count',
                     'CLV', 'Purchase_Frequency', 'Recency', 'Frequency', 'Monetary',
                     'R_Score', 'F_Score', 'M_Score', 'RFM_Score']

for col in numerical_features:
    if col in df_ml.columns:
        feature_columns.append(col)

# Encoded categorical features
for col in categorical_cols:
    if f'{col}_encoded' in df_ml.columns:
        feature_columns.append(f'{col}_encoded')

print(f"\n2. Selected Features for ML: {len(feature_columns)} features")
print(f"   Features: {', '.join(feature_columns[:10])}..." if len(feature_columns) > 10 else f"   Features: {', '.join(feature_columns)}")

# 3. Create final ML-ready dataset
X_ml = df_ml[feature_columns].copy()
X_ml = X_ml.dropna()

print(f"\n3. Final ML Dataset:")
print(f"   Shape: {X_ml.shape}")
print(f"   Features: {X_ml.shape[1]}")
print(f"   Samples: {X_ml.shape[0]}")

print("\n" + "=" * 70)
print("✓ Data preprocessing for ML complete!")
print("=" * 70)

# Display sample of processed data
print("\nSample of ML-ready data:")
X_ml.head()
````
